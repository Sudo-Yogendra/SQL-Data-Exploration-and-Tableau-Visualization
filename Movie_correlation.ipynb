```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
plt.style.use('ggplot')
import numpy as np
from matplotlib.pyplot import figure

%matplotlib inline
```

```python
plt.figure(figsize=(12,8))
```

```python
df = pd.read_csv('C:\Data Analysis\movies.csv')
```

```python
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print(f'{col} {pct_missing} %')
```

```python
df.dtypes
```

```python
df = df.fillna(0)
```

```python
df['budget'] = df['budget'].astype('int64')
```

```python
df['gross'] = df['gross'].astype('int64')
```

```python
df.head()
```

```python
# df['yearcorrect'] = df['released'].astype('str').split(',')[1].str[:4]
```

```python
df.sort_values(by = ['gross'], inplace=False, ascending=False)
```

```python
# drop duplicates
df.drop_duplicates()
```

```python
plt.figure(figsize=(10,8))
plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Gross Vs Budget')
plt.ylabel('budget of the movie')
plt.xlabel('Gross earnings of the movie')
plt.show()
```

```python
#using seaborn
plt.figure(figsize=(10,8))
sns.regplot(x='budget', y='gross', data = df, line_kws = {'color':'blue'})
plt.show()
```

```python
#correlation
```

```python
df.corr()
```

```python
plt.figure(figsize=(10,8))
correlation_matrix = df.corr(method='pearson')
sns.heatmap(correlation_matrix, annot=True)
plt.show()
```

```python
df_numerized = df
for colname in df_numerized.columns:
    if(df_numerized[colname].dtype == 'object'):
        df_numerized[colname] = df_numerized[colname].astype('category')
        df_numerized[colname] = df_numerized[colname].cat.codes
df_numerized
```

```python
plt.figure(figsize=(15,12))
correlation_matrix = df_numerized.corr(method='pearson')
sns.heatmap(correlation_matrix, annot=True)
plt.show()
```
